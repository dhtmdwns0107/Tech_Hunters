### 서버에서 Selenium 사용하기
- 크롬 브라우저 설치
    - sudo atp-get install -y chromium-browser
    - (버전확인) chromium-browser -version
- 크롬 드라이버 다운로드
    - wget https://chromedriver.storage.googleapis.com/78.0.3904.105/chromedriver_linux64.zip 
    - 개발자도구에서 링크가 확인되는데 /앞에 도메인이 생략된거라서 복사붙여넣기 했다.
    - sudo apt install unzip
    - unzip
    - chromedriver를 글로벌한 환경에서 실행시킬 수 있도록 파일이동
    - sudo mv chromedriver /usr/local/bin (move)
    - rm chromedriver_linux64.zip (remove)
- Selenium 패키지 설치
    - pip install selenium
    - pip install pillow
- 한글 폰트 설치
    - sudo apt-get install -y fonts-nanum*

```python
from selenium import webdriver
from PIL import Image as pil
```

```python
# headless 써야함
options = webdriver.ChromeOptions()
options.add_argument("headless")
options.add_argument("window-size=1920x1080") #캡처되는 크기
driver = webdriver.Chrome(options=options)
```

```python
driver.get("https://naver.com")
driver.save_screenshot("naver_main.png")
driver.quit()
```

```python
# 저장한 이미지 확인
pil_im = pil.open("naver_main.png")
pil_im
```
